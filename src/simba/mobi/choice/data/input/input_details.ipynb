## Decisions

- `hh_size` already captured by `single_household` and `nb_of_children_not_NA` -> removed from model
- `identified_as_male` not used in SBB models and does not seem very significant anyway -> removed from model
- `work_typology_urban` and `work_typology_rural` too correlated -> drop rural as reference category with intermediate
- `pt_travel_times` highly correlated with `home_distance_car` -> distance removed from model
- all `X_NA` pt travel times related variables and `work_typology_NA` highly correlated -> grouped into one variable
- `work_parking_NA` correlated with `work_time_flexibility_NA` -> removed from model
- `car_avail_NA` correlated with `has_driving_license_not_NA` -> removed from model
- `business_sector_agriculture_21` is only 0 -> removed from model
