```python
data_folder = '/media/nicolas-salvade/Windows/Users/DAF1/OneDrive - University College London/Documents/PhD - UCL/choice_set_location_travelmode/Data/mtmc/'
#data_folder = '../../../../../../../../choice_set_location_travelmode/Data/mtmc/'
individuals_path = data_folder + 'target_indiv.csv'
households_path = data_folder + 'household.csv'
```

```python
df_p = pd.read_csv(individuals_path)
df_h = pd.read_csv(households_path)
```

```python
#keeping only individuals who have answered the work_home question
df_p = df_p[df_p['work_home'] > 0]
```

```python
df_p.describe()
```

```python
#dropping columns with too many missing values and not usefull for the analysis
to_drop = ['learning_car', 'learning_moto', 'park_edu_car', 'use_park_edu', 'park_edu_cost', 'park_edu_moto', 'park_edu_bike', 'empl_status_3lev', 'edu_x', 'edu_y', 'edu_x_CH1903', 'edu_y_CH1903', 'edu_country', 'edu_lang', 'edu_region', 'edu_canton', 'edu_npa', 'edu_agglo2012', 'edu_agglo_groesse2012', 'edu_urban_char_2012', 'edu_stat_city_2012']

df_p = df_p.drop(to_drop, axis=1)
```

```python
#re[placing -99 by NaN
#df_p.replace(-99, np.nan, inplace=True)
```

```python
#if no or not known driving license, then no analysis on car use
df_p['dl_car'] = df_p['dl_car'].replace(2, 0)
df_p['dl_car'] = df_p['dl_car'].replace(-99, 0)
df_p['dl_moto'] = df_p['dl_moto'].replace(2, 0)
df_p['dl_moto'] = df_p['dl_moto'].replace(-99, 0)
```

```python
df_p.loc[df_p['dl_car']==0, 'park_work_car'] = 3
df_p.loc[df_p['dl_moto']==0, 'park_work_moto'] = 2
```

```python
#assigning never use park at work and free parking if the individual does not have a car spot at work
df_p.loc[df_p['park_work_car']==3, 'use_park_work'] = 4
df_p.loc[df_p['park_work_car']==3, 'park_work_cost'] = 0
#aasigning 0 to park_work_cost if the individual have a free car spot at work
df_p.loc[df_p['park_work_car']==1, 'park_work_cost'] = 0
```

```python
#create dummies for economic sectors

df_p.loc[:,'business_sector_agriculture'] = (df_p['econ_sector'] == 1).apply(int)
df_p.loc[:,'business_sector_industry'] = (df_p['econ_sector'] > 1).apply(int)
df_p.loc[:,'business_sector_services'] = (df_p['econ_sector_services'] > 0).apply(int)

to_drop = ['econ_sector', 'econ_sector_services']
df_p = df_p.drop(to_drop, axis=1)
```

```python
#create dummies for work schedule
df_p.loc[:, 'fixed_work_sched'] = (df_p['flex_work_sched'] == 1).apply(int)
df_p.loc[:, 'semi_flexible_work_sched'] = df_p['flex_work_sched'].apply(lambda x: 1 if x == 2 or x == 3 else 0)
df_p.loc[:, 'flex_work_sched'] = (df_p['flex_work_sched'] == 4).apply(int)

to_drop = ['flex_work_sched']
df_p = df_p.drop(to_drop, axis=1)
```

```python
#create dummies for pt subscription

df_p.loc[:, 'ptsub_ga'] = (df_p['pt_ga'] == 1).apply(int)
df_p.loc[:, 'ptsub_regional'] = df_p.apply(lambda x: 1 if x.pt_regional == 1 and x.pt_ga != 1 else 0, axis=1)
df_p.loc[:, 'ptsub_parcours'] = df_p.apply(lambda x: 1 if x.pt_parcours == 1 and x.pt_ga != 1 and x.pt_regional != 1 else 0, axis=1)
df_p.loc[:, 'ptsub_other'] = df_p.apply(lambda x: 1 if x.pt_other == 1 and x.pt_ga != 1 and x.pt_regional != 1 and x.pt_parcours != 1 else 0, axis=1)
df_p.loc[:, 'ptsub_junior'] = df_p.apply(lambda x: 1 if x.pt_junior == 1 and x.pt_ga != 1 and x.pt_regional != 1 and x.pt_parcours != 1 and x.pt_other != 1 else 0, axis=1)
df_p.loc[:, 'ptsub_ht'] = df_p.apply(lambda x: 1 if x.pt_ht == 1 and x.pt_ga != 1 else 0, axis=1)
df_p.loc[:, 'ptsub_track_sev'] = df_p.apply(lambda x: 1 if x.pt_track_sev == 1 else 0, axis=1)
df_p.loc[:, 'ptsub_ga_first'] = df_p.apply(lambda x: 1 if x.ga_first_second == 1 and x.ptsub_ga == 1 else 0, axis=1)
df_p.loc[:, 'ptsub_ga_second'] = df_p.apply(lambda x: 1 if x.ga_first_second == 2 and x.ptsub_ga == 1 else 0, axis=1)
df_p.loc[:, 'ptsub_parcours_first'] = df_p.apply(lambda x: 1 if x.parcours_first_second == 1 and x.ptsub_parcours == 1 else 0, axis=1)
df_p.loc[:, 'ptsub_parcours_second'] = df_p.apply(lambda x: 1 if x.parcours_first_second == 2 and x.ptsub_parcours == 1 else 0, axis=1)
df_p.loc[:, 'ptsub_regional_first'] = df_p.apply(lambda x: 1 if x.regional_first_second == 1 and x.ptsub_regional == 1 else 0, axis=1)
df_p.loc[:, 'ptsub_regional_second'] = df_p.apply(lambda x: 1 if x.regional_first_second == 2 and x.ptsub_regional == 1 else 0, axis=1)

to_drop = ['pt_ga', 'pt_regional', 'pt_parcours', 'pt_other', 'pt_junior', 'pt_ht', 'pt_track_sev', 'ga_first_second', 'parcours_first_second', 'regional_first_second']

df_p = df_p.drop(to_drop, axis=1)
```

```python
#pt employment subsidies dummy
df_p.loc[:, 'pt_emp_subsidies'] = df_p['pt_emp_subsidy'].apply(lambda x: 1 if x == 1 else 0)

#can walk alone dummy, it is assumed that if they cannot walk more than 200m alone it would have the same inpact on their transportation than if they couldn't walk
df_p.loc[:, 'can_walk_alone'] = df_p['max_walk_dist'].apply(lambda x: 1 if x == 1 else 0)

#wheelchair dummy
df_p.loc[:, 'use_wheelchair'] = df_p['wheelchair'].apply(lambda x: 1 if x == 1 else 0)

to_drop = ['wheelchair', 'max_walk_dist', 'pt_emp_subsidy']
```

```python
#mode availability dummies
df_p.loc[:, 'car_sharing'] = df_p['car_sharing'].apply(lambda x: 1 if x == 1 else 0)
```
